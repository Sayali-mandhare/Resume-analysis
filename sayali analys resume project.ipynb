```python
import PyPDF2
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def extract_text_from_pdf(file_path):
    pdf = PyPDF2.PdfReader(file_path)
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
    return text
```

```python
def clean_text(text):
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9 ]', '', text)
    return text.lower()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def get_similarity(resume_text, jd_text):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([resume_text, jd_text])
    similarity = cosine_similarity(vectors[0:1], vectors[1:2])
    return round(similarity[0][0] * 100, 2)
```

```python
resume_path = "resume_ analyzer project"

resume_text = clean_text(extract_text_from_pdf(resume_path))

jd_text = """
We are looking for a Data Analyst skilled in Python, SQL, Excel, and Data Visualization (Power BI or Tableau).
Experience with Machine Learning is a plus.
"""
jd_text = clean_text(jd_text)

match_score = get_similarity(resume_text, jd_text)
print(f"✅ Resume matches {match_score}% with the job description!")
```

```python
def find_missing_skills(resume_text, jd_text):
    resume_words = set(resume_text.split())
    jd_words = set(jd_text.split())
    missing = jd_words - resume_words
    return [word for word in missing if len(word) > 3]

missing_skills = find_missing_skills(resume_text, jd_text)
print("🧩 Missing Skills:", missing_skills)
```

```python
match_score = get_similarity(resume_text, jd_text)
print(f"✅ Resume matches {match_score}% with the job description!")
```
